```python
import pandas as pd
import shutil
import os
import gc
```

```python
df = pd.read_csv('validation.csv')
df
```

```python
# 创建存放验证集EEG文件的文件夹，如果不存在的话
validation_dir = 'validation_eegs'
if not os.path.exists(validation_dir):
    os.makedirs(validation_dir)
```

```python
# 遍历DataFrame中的egg_id
for egg_id in df.eeg_id:
    # 构建原始文件的完整路径
    file_path = os.path.join('E:/HMS2024/train_eegs/', f'{egg_id}.parquet')
    
    # 检查文件是否存在
    if os.path.exists(file_path):
        # 构建目标文件的完整路径
        destination_path = os.path.join(validation_dir, f'{egg_id}.parquet')
        # 复制文件
        shutil.copy(file_path, destination_path)
    else:
        print(f'File {file_path}.parquet not found.')

print("Files have been copied to the 'validation_eegs' folder.")
gc.collect()
```
